## عائلات المعالجات المتقاعدة
- **Canary**: ضمّت عائلة Canary تصاميم صغيرة تحتوي على ما بين 5 و16 كيوبت.
- **Falcon**: قدّمت عائلة أجهزة Falcon منصةً قيّمة للدوائر متوسطة الحجم، كما كانت بيئةً ممتازة لإثبات تحسينات الأداء وقابلية التوسع قبل نقلها إلى الأجهزة الأكبر حجمًا.
- **Egret**: أدخلت Egret ابتكارات المقرنات القابلة للضبط على منصة مكوّنة من 33 كيوبت، مما أسفر عن بوابات ثنائية الكيوبت أسرع وأعلى دقةً.
- **Hummingbird**: باستخدام تخطيط الكيوبتات السداسي الثقيل، أتاحت عائلة Hummingbird استيعاب ما يصل إلى 65 كيوبت.
- **Osprey**: كانت Osprey تقريبًا أربعة أضعاف حجم Eagle بـ433 كيوبت. استلزمت أحجام الشرائح الأكبر تحسينات إضافية في التغليف، إضافةً إلى كبلات مرنة مخصصة داخل التبريد التدريجي لاستيعاب متطلبات الإدخال/الإخراج الأكبر ضمن نفس البصمة السلكية.
- **Eagle**: دمجت عائلة معالجات Eagle تقنيات تغليف أكثر قابليةً للتوسع مقارنةً بالأجيال السابقة. تحديدًا، مرّت الإشارات عبر طبقات متعددة من الشرائح مما أتاح كثافة عالية للإدخال/الإخراج دون التضحية بالأداء.
